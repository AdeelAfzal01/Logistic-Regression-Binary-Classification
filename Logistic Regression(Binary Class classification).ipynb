# Data Preprocessing

### Dummies varialbes of catagorical vaiables

```python
sal_dummies = pd.get_dummies(ex_df.salary)
sal_dummies
```

```python
dep_dummies = pd.get_dummies(ex_df.Department)
```

```python
merged = pd.concat([ex_df,sal_dummies,dep_dummies], axis='columns')
merged
```

```python
X = merged.drop(['salary','Department','management','low'], axis='columns')
Y = merged.left
```

### Split the data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
ex_model = LogisticRegression()
```

# Training

```python
ex_model.fit(X_train,Y_train)
```

# Prediciton

```python
ex_model.predict(X_test)
```

# Accuracy

```python
ex_model.score(X_test,Y_test)
```
